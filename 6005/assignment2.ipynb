#### (a) Obtain the regression model of expend against the predictors, ratio, salary,takers, and total.

```python
import statsmodels.formula.api as smf

model = smf.ols('expend ~ ratio+salary+takers+total', data = data).fit() # Fit the model
print(model.summary())
```

#### (b) By considering a 50%-50% split of the data into training set and test set, output the root mean squared prediction error of the following models
`A, Expend against ratio, salary, takers, and total`

`B, Expend against ratio, salary, takers, verbal, and math`

`C, Expend against ratio, salary, takers, verbal`

`D, Expend against ratio, salary, takers, and math`

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# A, Expend against ratio, salary, takers, and total
X_train, X_test, y_train, y_test = train_test_split(data[['ratio', 'salary','takers','total']], data['expend'], test_size=0.5, random_state=42)
model = smf.ols('expend ~ ratio+salary+takers+total', data = data).fit() 
pre_val = model.predict(data)
act_val = data['expend']
bias = np.mean(pre_val - act_val)
rmse = np.sqrt(mean_squared_error(act_val, pre_val))

print("Bias:", bias)
print("RMSE:", rmse)
```

```python
# B, Expend against ratio, salary, takers, verbal, and math
X_train, X_test, y_train, y_test = train_test_split(data[['ratio', 'salary','takers','verbal','total']], data['expend'], test_size=0.5, random_state=42)
model = smf.ols('expend ~ ratio+salary+takers+verbal+total', data = data).fit() 
pre_val = model.predict(data)
act_val = data['expend']
bias = np.mean(pre_val - act_val)
rmse = np.sqrt(mean_squared_error(act_val, pre_val))

print("Bias:", bias)
print("RMSE:", rmse)
```

```python
# C, Expend against ratio, salary, takers, verbal
X_train, X_test, y_train, y_test = train_test_split(data[['ratio', 'salary','takers','verbal']], data['expend'], test_size=0.5, random_state=42)
model = smf.ols('expend ~ ratio+salary+takers+verbal', data = data).fit() 
pre_val = model.predict(data)
act_val = data['expend']
bias = np.mean(pre_val - act_val)
rmse = np.sqrt(mean_squared_error(act_val, pre_val))

print("Bias:", bias)
print("RMSE:", rmse)
```

```python
# D, Expend against ratio, salary, takers, and math
X_train, X_test, y_train, y_test = train_test_split(data[['ratio', 'salary','takers','math']], data['expend'], test_size=0.5, random_state=42)
model = smf.ols('expend ~ ratio+salary+takers+math', data = data).fit() 
pre_val = model.predict(data)
act_val = data['expend']
bias = np.mean(pre_val - act_val)
rmse = np.sqrt(mean_squared_error(act_val, pre_val))

print("Bias:", bias)
print("RMSE:", rmse)
```

## Question 2. The data file bank.csv contains a subset of a huge credit dataset of a German bank. This dataset has 1000 observations and 10 columns:

#### (a) Obtain the regression coefficients in the logistic regression model from the data using Python.

```python
path =  './Data/bank.csv'
data = pd.read_csv(path, header=0)
data
```

```python
import statsmodels.formula.api as smf
from sklearn.metrics import confusion_matrix

data['Class'] = data['Class'].astype('category')
data['Class'] = data['Class'].cat.codes

model = smf.mnlogit('Class ~ Check+Duration+Credit+Purpose+Amount+Saving+Employ+Install+Marriage', data=data).fit()

model.summary()
```

#### (b) Predict the probability of “Class = Good” for the new observation with Check = 100 DM, Duration = 36 months, No credit taken, Business purpose, Amount = 1953 DM, Saving = 50 DM, Employed for 10 years, Install rate = 4%, Male single.

```python
newdata = pd.DataFrame({
    'Check': [100],
    'Duration':[36],
    'Credit': [0],  
    'Purpose': [1], 
    'Amount': [1953],
    'Saving': [50],
    'Employ': [10],
    'Install':[4],
    'Marriage': [3],  
})

pred_probs = model.predict(newdata)
print(pred_probs)

pred_classes = pred_probs.idxmax(axis=1)
print(pred_classes)
```
