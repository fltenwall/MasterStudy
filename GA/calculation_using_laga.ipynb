```python
import numpy as np
import pandas as pd
from LSGA_result import best_ind_i, best_ind_j, best_ind_k
```

```python
# Combine indices into a DataFrame
A = pd.DataFrame({'i_ind': best_ind_i, 'j_ind': best_ind_j, 'k_ind': best_ind_k})
A
```

```python
# Generate random values
alpha = np.concatenate(([0], np.random.normal(size=99)))
beta = np.concatenate(([0], np.random.normal(size=99)))
gamma = np.concatenate(([0], np.random.normal(size=99)))
mu = np.random.normal()

# Create the 3D array Y
Y = mu + np.add.outer(np.add.outer(alpha, beta), gamma) + 0.01 * np.random.normal(size=(100, 100, 100))
```

```python
i_ind = A['i_ind'].values
j_ind = A['j_ind'].values
k_ind = A['k_ind'].values

# 使用這些索引從 Y 中選擇相應的值
matched_values = Y[i_ind, j_ind, k_ind]
matched_values_pd = pd.DataFrame({'y':matched_values})
YA = pd.concat([A, matched_values_pd], axis=1)
YA
```

```python
freq_i = np.bincount(i_ind, minlength=100)
freq_j = np.bincount(j_ind, minlength=100)
freq_k = np.bincount(k_ind, minlength=100)
print(f'freq_i: {freq_i}\n freq_j: {freq_j}\n freq_k: {freq_k}\n')
```

```python
diagonal_matrix_alpha_alpha = np.diag(freq_i)
diagonal_matrix_beta_beta = np.diag(freq_j)
diagonal_matrix_gamma_gamma = np.diag(freq_k)
freq = np.concatenate((freq_i, freq_j,freq_k), axis=0)
D = np.diag(freq)
D
```

```python
D_alpha_beta = np.empty((100,), dtype=object)

# 將每個元素初始化為空數組
for i in range(100):
    D_alpha_beta[i] = np.array([])

arr_ind = 0
for gamma in range(2,100):
    D_alpha_beta[arr_ind] = []
    for alpha in range(2,100):
        count = 0
        for beta in range(2,100):
            count += len(YA[(YA['k_ind'] == gamma) & (YA['i_ind'] == alpha) & (YA['j_ind'] == beta)] == True)
        D_alpha_beta[arr_ind].append(count)
    arr_ind += 1
```
