```python
import pandas as pd

path = './Residential-Building-Data-Set.xlsx'
origin_data = pd.read_excel(path)
origin_data_columns_length = len(origin_data.columns)
count = 0
index_label = True
first_level = []
iterables = []

while count < origin_data_columns_length:
    name = origin_data.columns[count]
    if name.find('Unnamed') == -1:
        first_level.append(name)
        index_label = True
    else:
        if index_label == True:
            iterables.append([])
            iterables[-1].append(name)
        else:
            iterables[-1].append(name)
        index_label = False
    count = count + 1

colunms = [(name, 'v'+str(order+1)) for (index, name) in enumerate(first_level) for (order, colunm) in enumerate(iterables[index])]
colunms
colunms = pd.MultiIndex.from_tuples(colunms)
pd.DataFrame(
        origin_data, 
        columns=colunms,
    )
```
