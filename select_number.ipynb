```python
import numpy as np

array_15_by_10_by_20 = np.random.uniform(0, 1, size=(1000, 1000, 1000))
 
print(array_15_by_10_by_20)
```

```python
item_count = 1000
select_list = []
count = 0
first_level_random_used = set()
second_level_random_used = set()

while True:
    first_level_random = np.random.randint(1, 1000)
    second_level_random = np.random.randint(1, 1000)
    
    while first_level_random not in first_level_random_used:
        first_level_random_used.add(first_level_random)
        break
    while second_level_random not in second_level_random_used:
        second_level_random_used.add(second_level_random)
        break
    
    for second_array in array_15_by_10_by_20[first_level_random]:
        second_array_mean = np.mean(array_15_by_10_by_20[second_level_random])
        for item in second_array:
            if item > second_array_mean:
                count = count + 1
                select_list.append(item)
            if count >= item_count:
                break
        if count >= item_count:
                break
    if count >= item_count:
                break
print(select_list)
print(len(select_list))
print(np.sum(np.array(select_list)))
```
